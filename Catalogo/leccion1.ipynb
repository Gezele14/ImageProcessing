* Sea $A\in\mathbb{R}^{m\times n}$, la descomposición en valores singulares $(SvD)$ es una factorización de la forma 
$$A = U\Sigma V^T,$$
    
* Donde $U\in\mathbb{R}^{m\times m}$ y $V\in\mathbb{R}^{n\times n}$ son matrices ortogonales, es decir:
$$U U^T = U^T U = Im $$
$$V V^T = V^T V = Im $$
    
Y $\Sigma\in\mathbb{R}^{m\times n}$ es una matriz diaginal generalizada con los valores singualres de la matriz


* Sea $A\in\mathbb{R}^{m\times n}$ y $K \leq min(m,n)$. Una imagen $A$ se puede comprimir con un porcentaje $\frac{k}{min(m,n)}100$ utilizando la siguiente fórmula:
$$A_{r} = U_{r} \Sigma_{r} V_{r}$$

Donde $U_{r}$ y $V_{r}$ están dadas por las primeras $k$ columnas de $U$ y $V$ respectivamente
