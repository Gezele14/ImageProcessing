* Sea $A\in\mathbb{R}^{m\times n}$, la $S,V,D$ de $A$ es la factorización de manera:
$$ A= U \sum V^{T} $$
    
* Aproximacion de $A$, utilizando $K$ valores singulares es:
    
$$A_k = U_k \sum V_k^{T} $$ 
$$U_k = [U_1, U_2, ..., U_k] $$
$$V_k = [V_1, V_2, ..., V_k] $$
    
* Para obtener la escala de grises de una imagen, se calcula el promedio de los colores de la imagen.

## Negativo de una imagen

* Sea $A\in\mathbb{R}^{m\times n}$ la representación en formato "double" de una imagen. El negativo de dicha imagen A, es la matriz $B\in\mathbb{R}^{m\times n}$, tal que:

$$B_{ij} = -A_{ij} + 1 $$

```octave
clc; clear

%Cargar la imagen
A = imread('files/cameraman.jpg');
A = im2double(A);
[m,n] = size(A);
s = svd(A); %Escribe los valores singulares de la matriz A en el vector s
r = min([m,n]);
%plot(1:r,s)

[U,S,V] = svd(A);
k = 150;
Uk = U(:,1:k); Sr = S(1:k,1:k); Vr = V (:,1:k);
A_k = Uk * Sr * Vr';

subplot(1,2,1)
imshow(A)
title('Imagen Original')
subplot(1,2,2)
imshow(A)
title('Imagen Reconstruida')
```

```octave
clc; clear

%Cargar la imagen
A = imread('files/cameraman.jpg');
A = im2double(A);
[m,n] = size(A);
s = svd(A); %Escribe los valores singulares de la matriz A en el vector s
r = min([m,n]);

[U,S,V] = svd(A);

dimensiones = [1 20:20:460 469];
for k = dimensiones
  Uk = U(:,1:k); Sr = S(1:k,1:k); Vr = V (:,1:k);
  A_k = Uk * Sr * Vr';
  error = norm(A-A_k);
  subplot(1,2,1)
  imshow(A)
  title('Imagen Original')
  subplot(1,2,2)
  imshow(A)
  title(['Imagen Reconstruida con k =' num2str(k) 'con error de ' num2str(error)])
  %pause(0.5)
end
```

```octave
clc; clear
pkg load image
%Cargar la imagen
A = imread('files/cameraman.jpg');
A = im2double(A);

B = imread('files/fondo.jpg');
B = im2double(B);

C = (1/3)*(B(:,:,1)+B(:,:,2)+B(:,:,3));
imshow(C)

A1 = imresize(A,[256,256]);
C1 = imresize(C,[256,256]);
subplot(1,3,1)
imshow(A1)
title('A1')
subplot(1,3,2)
imshow(C1)
title('C1')
D = (A1+C1)/2;
subplot(1,3,3)
imshow(D)
title('A1 + C1')
```
