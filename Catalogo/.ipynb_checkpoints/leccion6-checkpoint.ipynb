## Transformaciones en imágenes a escala de grises

* Con el fin principal de mejorar una imagen, por lo que se tienen dos objetivos:

    - Mejorar la calidad de una imagen desde el punto de vista humano.
    - Modificar una imagen para desarrollo de cierto tipo de análisis.
    
* Las transformaciones que se verán mas adelante modifican el valor de los pixeles, no su posición


* Transformación Puntual de una imagen en escala de grises


* Sea $A$ una imagen de tamaño $mxn$. Vamos a obtener otra imagen $B$ del mismo tamaño:


$$B(x,y) = T(A(x,y))$$


* Donde en $A(x,y)$ se transforma el valor del pixel.


* En general, la transformación será escrita de la siguiente forma:

$$s = T(r)$$

Donde

$$r = A(x,y)$$
$$s = B(x,y)$$ 

## Tipos de Transformaciones

## 1. Transformación Lineal

* La Transformación Lineal se escribe de la siguiente forma: 

$$T(r) = c r+b $$

Donde $c$ es la constante de la imagen y $b$ es el brillo de la imagen.

```octave
clc; clear;

A = imread('files/boat.jpg');

subplot(1, 2, 1)
imshow(A)
title('imagen original')

c = 2; 

for b = 0:20:200 
  A = c*A+b;
  subplot(1, 2, 2)
  imshow(A)
  title('imagen transformada')
  pause(0.25)
endfor
```

## 2. Transformación de AutoContraste

* La Transformación de AutoConstraste se realiza con la siguiente formula:


$$\begin{equation*}
T_{ac}(r) = \frac{L-1}{r_{max}-r_{min}} * (r-r_{min})
\end{equation*}$$



$$\begin{equation*}
T_{ac}(r) = c_{ac} *r - b_{ac}
\end{equation*}$$

Donde $L$ es la intensidad máxima de cualquier imagen en el formato de imagen. El valor de $r_{max}$ es el valor máximo de $A$ y $r_{min}$ es el valor mínimo.





```octave
clc; clear;
A = imread('files/boat.jpg');
subplot(1, 3, 1)
imshow(A)
title('imagen Original')

A = imread('files/boat_new.jpg');
subplot(1, 3, 2)
imshow(A)
title('imagen nueva')

L = 255;
rmax = max(max(A));
rmin = min(min(A));

A = (L-1) / (rmax-rmin) * (A-rmin);
subplot(1, 3, 3)
imshow(A)
title('imagen transformada')
```

## 4. Transformación Exponencial

* La Transformación Exponencial se realiza con la siguiente formula:

$$\begin{equation*}
T(r) = c * r^{\alpha}
\end{equation*}$$

Donde $c$ y $\alpha$ son constantes y $c$ es el paramtero de escala.

Además $\alpha$ define el tipo de brillo, donde si $\alpha > 1$ la imagen original se aclara, y si  $\alpha < 1$ la imagen se oscurece.


```octave
clc; clear;
A = imread('files/boat.jpg');
subplot(1, 2, 1)
imshow(A)
title('imagen Original')

c = 1;

for a = 0:0.1:2 
  B = c * power(A,a);
  subplot(1, 2, 2)
  imshow(B)
  title('imagen transformada')
  pause(0.25)
endfor
```

## 5. Transformación Logarítmica

* La Transformación Logarítmica se realiza con la siguiente formula: 

$$\begin{equation*}
T(r) = c * ln(1+r)
\end{equation*}$$

Donde $c$ es una constante.

<h2>Transformación por Partes</h2>
<p>La Transformación por partes está dada por</p>
\[   
T(r) = 
     \begin{cases}
       \text{$T_{1}(r)$,} &\quad\text{$r \epsilon A_{1}$}\\
       \text{$T_{2}(r)$,} &\quad\text{$r \epsilon A_{2}$}\\
       \text{$\vdots$} &\quad\text{$\vdots$}\\
       \text{$T_{n}(r)$,} &\quad\text{$r \epsilon A_{n}$}\\
     \end{cases}
\]

<p>Se realiza como ejercicio la siguiente transformación</p>
\[   
T(r) = 
     \begin{cases}
       \text{$255$} &\quad\text{si $r \geq 127$}\\
       \text{$0$} &\quad\text{si $r < 127$}\\
     \end{cases}
\]

```octave
clc; clear;

A = imread('files/boat.jpg');
subplot(1, 2, 1)
imshow(A)
title('Imagen Original')

[m,n,r] = size(A);
B = zeros(m,n,r);
B = uint8(B);
for x = 1:m
  for y = 1:n
    if A(x,y) >= 127
      B(x,y) = 255;
    else
      B(x,y) = 0;
    endif
  endfor
endfor

subplot(1, 2, 2)
imshow(B)
title('Imagen Transformada')
```
