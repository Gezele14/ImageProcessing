<h1 style="text-align: center;">Procesamiento de Imagenes con Histogramas</h1>

<h2>Histograma</h2>

Es el gráfico de la representación de distribuciones de frecuencias, en el que se emplean rectángulos dentro de unas coordenadas.

En nuestro caso, utilizaremos los histogramas que determinan el valor de la intensidad del pixel de una imagen.

Considere $A  \epsilon  R^{4 x 4}$, tal que:
<p>$$A = \begin{pmatrix}
0 & 5 & 3 & 1 \\ 
2 & 5 & 1 & 0 \\ 
0 & 2 & 4 & 1 \\ 
0 & 0 & 0 & 1
\end{pmatrix}$$</p>

Es claro que los valores que toma $A$ están en el conjunto {0, 1, 2, 3, 4, 5}

El histograma es el siguiente:

<img src="media/histograma.png"/>

<ul>
    
<li> En imágenes a escala de grises, tenemos una matriz $A  \epsilon  R^{m x n}$ donde cada entrada (pixel) toma los valores en el conjunto {0, ..., 255}. </li>

<li>El histograma de una imagen es la representación de la frecuencia de aparición de cada nivel de grises en la imagen.</li>

<li>Cada entrada individual del histograma se expresa en la siguiente manera:</li>
    
</ul>

<p>$$h(x) = hx = card\left\lbrace (x, y) / f(x, y) = k \right\rbrace$$</p>
<p>$$h = \left[ h_{(0)}, h_{(1)}, ..., h_{(255)} \right ]^{T}$$</p>

<h3>Código en Octave</h3>

```octave
clc; clear;
pkg load image
% Ejemplo 1 de histograma

A = imread('images/peppers.jpg');

subplot(2, 2, 1);
imshow(A);

% Forma 1 de calcular el histograma
h1 = zeros(256, 1);
[m, n] = size(A);

tic % Calcular el tiempo de la forma 1
for i = 1 : m
  for j = 1 : n
    h1(A(i, j) + 1) += 1;
  endfor
endfor

subplot(2, 2, 2);
bar(0:255, h1);
title('Forma 1');
xlim([0 255]);
t1 = toc


tic % Calcular el tiempo de la forma 2
% Forma 2 de calcular el histograma
h2 = zeros(256, 1);
for i = 0 : 255
  h2(i + 1) = sum(sum(A == i));
endfor

subplot(2, 2, 3);
bar(0:255, h2);
title('Forma 2');
xlim([0 255]);
t2 = toc

% Forma 3 de calcular el histograma
tic % Calcular el tiempo de la forma 3
subplot(2, 2, 4);
imhist(A) % Comando de Octave para calcular el histograma
title('Forma 3');
t3 = toc
```

<h2>Técnicas para Modificar Imágenes con el Histograma</h2>

<h3>1 - Ecualización del Histograma</h3>

Es una técnica que cambia la distribución de nivel de gris de una imagen de tal manera  que se obtiene un histograma resultante 
uniforme en el que el porcentaje de pixeles de cada nivel de grises es el mismo.

Para esto necesitamos una función auxiliar $T(r)$, que satisface 2 criterios:
<ul>
    <li>$T(r)$ es una función creciente [0, 255]</li>
    <li>$0 \leq T(r) \leq 1, \forall r \epsilon [0, 255]$</li>
</ul>

Un ejemplo de esta transformación es la función de distribución acumulada (cdf).
<p>$$\delta _{k} = T(r_{k}) = \sum_{j = 0}^{k} \frac{n_{j}}{n}$$</p>
donde $n$ es el número total de pixeles.

<h3>Código en Octave</h3>

```octave
clc; clear;
pkg load image
% Ejemplo 1 de histograma

A = imread('images/peppers.jpg');

subplot(1, 3, 1);
imshow(A);

[m, n] = size(A);

% Calcular el histograma
h2 = zeros(256, 1);
for i = 0 : 255
  h2(i + 1) = sum(sum(A == i));
endfor

subplot(1, 3, 2);
bar(0:255, h2);
title('Histograma');
xlim([0 255]);

% Distribucion acumulada
ac = zeros(256, 1);
for i = 0 : 255
  ac(i + 1) = sum(h2(1 : i + 1)) / (m * n);
endfor

subplot(1, 3, 3);
bar(0:255, ac);
title('Distribucion Acumulada');
xlim([0 255]);
```

<h3>Ejemplo</h3>

Sea 
<p>$$h = \left[ 1 \space 3 \space 2 \space 4\right ]$$</p>
El valor acumulada es
<p>$$a = \left[ \frac{1}{4} \space \frac{4}{4} \space \frac{6}{4}\space \frac{10}{4}\right ]$$</p>

<h3>Método</h3>

Sea $A  \epsilon  R^{m x n}$, los valores en sus entradas iguales a 0, 1, ..., 255.

Cambiamos el valor de la posición $A_{ij}$ por el valor $S_{k} * 255$, es decir, la transformación consiste en la siguiente fórmula:
<p>$$A_{ij} = \left[ S_{k} * 255 \right]$$</p>
donde $A_{ij} = k$

<h3>Código en Octave</h3>

```octave
clc; clear;
pkg load image
% Ejemplo 1 de histograma

A = imread('images/sydney.jpg');

subplot(2, 2, 1);
imshow(A);

[m, n] = size(A);

% Calcular el histograma
h2 = zeros(256, 1);
for i = 0 : 255
  h2(i + 1) = sum(sum(A == i));
endfor

subplot(2, 2, 2);
bar(0:255, h2);
title('Histograma');
xlim([0 255]);

% Distribucion acumulada
ac = zeros(256, 1);
for i = 0 : 255
  ac(i + 1) = sum(h2(1 : i + 1)) / (m * n);
endfor

% Obtener la nueva imagen aplicando la tecnica de ecualizacion
B = zeros(m, n);
B = uint8(B);

for i = 1 : m
  for j = 1 : n
    B(i, j) = round(ac(A(i, j) + 1) * 255); 
  endfor
endfor

subplot(2, 2, 3);
imshow(B);

% Calcular el histograma
h3 = zeros(256, 1);
for i = 0 : 255
  h3(i + 1) = sum(sum(B == i));
endfor

subplot(2, 2, 4);
bar(0:255, h3);
title('Histograma');
xlim([0 255]);
```

<h3>2 - Estiramiento del Histograma</h3>

Esta técnica consiste en una transformación lineal que expanda parte del histograma original tal que la intensidad original que está en el intervalo $\left[ r_{min}, r_{max}\right[$, ocupe toda la escala, $\left[ 0, 255 \right]$.

<img src="media/estiramiento.png"/>

Matématicamente, cada pixel de la nueva imagen $B$ se obtiene de la fórmula:
<p>$$B_{ij} = \frac{A_{ij}-r_{min}}{r_{max}-r_{min}}*255 = \frac{255}{r_{max}-r_{min}}*(A_{ij}-r_{min})$$</p>
donde $r_{min}$ y $r_{max}$ son los valores más pequeños y más grande en escala de grises de $A$.

<h3>Código en Octave</h3>

```octave
clc; clear;
pkg load image
% Ejemplo 1 de histograma

A = imread('images/sydney.jpg');

subplot(2, 2, 1);
imshow(A);

[m, n] = size(A);

% Calcular el histograma
h2 = zeros(256, 1);
for i = 0 : 255
  h2(i + 1) = sum(sum(A == i));
endfor

subplot(2, 2, 2);
bar(0:255, h2);
title('Histograma');
xlim([0 255]);

% Modificacion de la Imagen, Estirando el  Histograma
r_min = min(min(A)); r_max = max(max(A));
B = uint8(zeros(m, n));
B = (255 / (r_max - r_min)) * (A - r_min);

subplot(2, 2, 3);
imshow(B);

% Calcular el histograma
h3 = zeros(256, 1);
for i = 0 : 255
  h3(i + 1) = sum(sum(B == i));
endfor

subplot(2, 2, 4);
bar(0:255, h3);
title('Histograma Estirado');
xlim([0 255]);
```

<h3>3 - Reducción del Histograma</h3>

Esta técnia modifica el histograma original de tal manera que comprime la dinámica de escalad de grises $\left[ r_{min}, r_{max} \right]$ a otra escala $\left[ S_{min}, S_{max} \right]$, donde:
<p>$$r_{min} < S_{min}$$ y $$S_{max} < r_{max}$$</p>

<img src="media/reduccion.png" />

Metemáticamente, cada pixel $B_{ij}$ de la imagen B modificado se obtiene de la fórmula:
<p>$$B_{ij} = \left [ \frac{S_{max}-S_{min}}{r_{max}-r_{min}} \right ]*(A_{ij}-r_{min}) + S_{min}$$</p>

<h3>Código en Octave</h3>

```octave
clc; clear;
pkg load image
% Ejemplo 1 de histograma

A = imread('images/peppers.jpg');

subplot(2, 2, 1);
imshow(A);

A = double(A);
[m, n] = size(A);


% Calcular el histograma
h2 = zeros(256, 1);
for i = 0 : 255
  h2(i + 1) = sum(sum(A == i));
endfor

subplot(2, 2, 2);
bar(0:255, h2);
title('Histograma');
xlim([0 255]);

% Modificacion de la Imagen, Reduccion el  Histograma
r_min = min(min(A)); %=0
r_max = max(max(A)); %227
s_min = 50;
s_max = 200;


B = zeros(m,n);
B = round(((s_max-s_min)/(r_max-r_min))*(A-r_min)+s_min);
B = uint8(B);

subplot(2, 2, 3);
imshow(B);

% Calcular el histograma
h3 = zeros(256, 1);
for i = 0 : 255
  h3(i + 1) = sum(sum(B == i));
endfor

subplot(2, 2, 4);
bar(0:255, h3);
title('Histograma');
xlim([0 255]);
```
